Ahora vamos a analizar las distintas páginas web que derivan de la principal.

Nos podemos encontrar dos tipos:

El primer tipo son webs a las que se le hace referencia directamente desde su dirección absoluto y el segundo tipo se le hace referencia en relacción a la web principal.

```python
direccionesinternas=[]
direccioneshttps=[]
for a in soup.findAll(href=True):
    if a['href']:
        print(a['href'])
        if ('https' in a['href']):
            if (a['href'] not in direccioneshttps):
                direccioneshttps.append(a['href'])
        elif(a['href'][0]=="/"):
            if (a['href'] not in direccionesinternas):
                direccionesinternas.append(a['href'])
```

Una vez obtenidas todas las direcciones vamos a unificarlas todas en un mismo estilo que va a ser el de la dirección entera

```python
for url in direccionesinternas:
    url=Url+url
    direccioneshttps.append(url)
```

```python
#Presencial
reqP = requests.get("https://www.uax.com/presencial")
soupP = BeautifulSoup(reqP.text)
```

```python
for link in soupP.findAll(href=True):
    if 'titulaciones' in link['href']:
        print(link['href'])
```

Como podemos ver solo aparecen algunas titulaciones y no todas, esto es debido a que están separadas en páginas.

```python
titulaciones=[]
for i in range(1,7):
    print("https://www.uax.com/presencial?page="+str(i))
    reqP = requests.get("https://www.uax.com/presencial?page="+str(i))
    soupP = BeautifulSoup(reqP.text)
    for link in soupP.findAll(href=True):
        if (('titulaciones' in link['href']) & (link['href'] not in titulaciones)):
            titulaciones.append(link['href'])
```

```python
titulaciones
```

```python
reqP6 = requests.get("https://www.uax.com/presencial?page=6")
soupP6 = BeautifulSoup(reqP6.text)
for link in soupP6.findAll(href=True):
    if 'titulaciones' in link['href']:
        print(link['href'])
```
